```python
#Python function that returns length of longest word:
    
def get_longest_word(word_list):
    max_len = 0
    for word in word_list:
        if len(word) > max_len:
            max_len = len(word)
            result = word   
    return max_len, result

get_longest_word(['one','two','three','four','five','six','seven','eight','nine','ten','eleven'])


```

```python
#Count number of unique characters in a string

def count_char(string):
    return len(set(list(string)))

count_char('characters')
```

```python
#Sort ascending and descending in dictionary by value

dict1={'Blue':5,'Red':8,'Yellow':2,'Purple':12,'Green':11}

sorted_d1 = sorted(dict1.items(), key=lambda x:x[1])
sorted_d2 = sorted(dict1.items(), key=lambda x:x[1],reverse=True)

print('Ascending:', sorted_d1)
print('Descending:', sorted_d2)
```

```python
#Change given string to a new string where the first and last chars have been exhanged

def exchange(string):
    return (string[-1]+string[1:-1]+string[0])

exchange('python')
```

```python
#Return True if two lists have at least one common member

recipe={'chicken','broth','noodles','carrots','celery','butter','salt'}
pantry={'salt','noodles','onion','milk','crackers','beans','peas'}

def common_members(list1,list2):
    set1 = set(list1)
    set2 = set(list2)
    if (set1 & set2):
        return True
    else:
        return False
    
common_members(recipe,pantry)
```

```python
import sys 

def minCoins(coins, m, V): 
 
    if (V == 0): 
        return 0

    res = sys.maxsize 

    for i in range(0, m): 
        if (coins[i] <= V): 
            sub_res = minCoins(coins, m, V-coins[i]) 
            
            if (sub_res != sys.maxsize and sub_res + 0.01 < res): 
                res = sub_res + 0.01

    return res 

coins = [1, 0.25, 0.10, 0.01] 
m = len(coins) 
V = 1.11
print("Minimum coins:",minCoins(coins, m, V)) 
```
